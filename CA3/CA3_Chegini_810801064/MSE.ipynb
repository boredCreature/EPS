# Linear Regression

Linear regression is a fundamental statistical method used to model the relationship between a dependent variable and one or more independent variables. The goal is to find the best-fitting line through the data points.

The linear regression model can be expressed as:

$$ y = f(x) = a x + b$$

Where:
- $ y $ is the dependent variable.
- $ x $ is the independent variable.
- $ b $ is the y-intercept.
- $ a $ is the slope of the line.

We utilize the Mean Squared Error (MSE) approach to determine the coefficients $a$ and $b$ based on a set of sample data. 
$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - f(x_i))^2 $$

Where:
- $ n $ is the number of observations.
- $(x_i, y_i)$ is a given sample data 

## Finding the Solution

To find the solution for the coefficients $ b $ and $ a $, we can use calculus. The steps are as follows:

1. **Set up the error function**:
   $$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - (a x_i + b))^2 $$

2. **Derive the  solution** for $ b $ and $ a $ by taking the partial derivatives of the error function $ MSE $ and setting them to zero. Then solve the equations to find $ b $ and $ a $.

   ######################  TODO  ########################
   ### Step 1: Take the Partial Derivative of MSE with Respect to \( a \)
To minimize the MSE, we take the partial derivative with respect to \( a \):

$$
\frac{\partial MSE}{\partial a} = -\frac{2}{n} \sum_{i=1}^{n} x_i \left( y_i - (a x_i + b) \right)
$$

Set the derivative equal to zero to find the optimal value of \( a \):

$$
\sum_{i=1}^{n} x_i \left( y_i - a x_i - b \right) = 0
$$

### Step 2: Take the Partial Derivative of MSE with Respect to \( b \)
Next, take the partial derivative of MSE with respect to \( b \):

$$
\frac{\partial MSE}{\partial b} = -\frac{2}{n} \sum_{i=1}^{n} \left( y_i - (a x_i + b) \right)
$$

Set this derivative equal to zero to find the optimal value of \( b \):

$$
\sum_{i=1}^{n} \left( y_i - a x_i - b \right) = 0
$$

### Step 3: Solve the System of Equations
We now have the system of equations:

1. $$ \sum_{i=1}^{n} x_i \left( y_i - a x_i - b \right) = 0 $$
2. $$ \sum_{i=1}^{n} \left( y_i - a x_i - b \right) = 0 $$

Solve these two equations simultaneously to find the values of \( a \) and \( b \).

### Step 4: Solve for the Slope \( a \)
From the first equation, we can solve for \( a \):

$$
a = \frac{n \sum_{i=1}^{n} x_i y_i - \sum_{i=1}^{n} x_i \sum_{i=1}^{n} y_i}{n \sum_{i=1}^{n} x_i^2 - \left( \sum_{i=1}^{n} x_i \right)^2}
$$

### Step 5: Solve for the Intercept \( b \)
From the second equation, we can solve for \( b \):

$$
b = \frac{1}{n} \left( \sum_{i=1}^{n} y_i - a \sum_{i=1}^{n} x_i \right)
$$

### Step 6: Final Formulas for \( a \) and \( b \)
Thus, the final formulas for the slope \( a \) and intercept \( b \) are:

$$
a = \frac{n \sum_{i=1}^{n} x_i y_i - \sum_{i=1}^{n} x_i \sum_{i=1}^{n} y_i}{n \sum_{i=1}^{n} x_i^2 - \left( \sum_{i=1}^{n} x_i \right)^2}
$$

$$
b = \frac{1}{n} \left( \sum_{i=1}^{n} y_i - a \sum_{i=1}^{n} x_i \right)
$$

These are the formulas for the slope and intercept in simple linear regression obtained by minimizing the MSE.

```python
class SimpleLinearRegression:
    """
    A class to represent a simple linear regression model.

    Attributes:
        slope (float): The slope of the regression line.
        intercept (float): The intercept of the regression line.
    """
    def __init__(self):
        self.slope: float = 0  # Slope of the regression line
        self.intercept: float = 0  # Intercept of the regression line

    def fit(self, x: list[float], y: list[float]) -> None:
        """
        Fit the linear regression model to the data.

        Args:
            x (list[float]): List of independent variable values.
            y (list[float]): List of dependent variable values.
        """
        ######################  TODO  ########################
        # Calculate slope and intercept based on obtained formula
        n = len(x)

        sum_x = sum(x)
        sum_y = sum(y)
        sum_xy = sum(xi * yi for xi, yi in zip(x, y))
        sum_x_squared = sum(xi ** 2 for xi in x)

        self.slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
        self.intercept = (sum_y - self.slope * sum_x) / n
        ######################  TODO  ########################


    def predict(self, x: list[float]) -> list[float]:
        """
        Predict the dependent variable values for given independent variable values.

        Args:
            x (list[float]): List of independent variable values to predict.
        
        """
        ######################  TODO  ########################
        # Use the formula y = b + m * x to calculate predictions
        return [self.intercept + self.slope * xi for xi in x]
        ######################  TODO  ########################
```

## Prove the Solution

### Linear Regression Model

$$
y = f(x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \ldots + \theta_n x_n
$$

Where:
- $ \theta $ is the vector of parameters (weights),
- $ x $ is the feature vector.

We utilize the Mean Squared Error (MSE) approach to determine the coefficient $\theta$ based on a set of sample data. 

$$
MSE = \frac{1}{m} \sum_{i=1}^{m} (y_i - f(x))^2
$$

Where:
- $ m $ is the number of samples,
- $(x_i, y_i)$ is a given sample data.

### Matrix Notation

In matrix form, the cost function can be expressed as:

$$
MSE = \frac{1}{m} (y - x\theta)^T (y - x\theta)
$$

### Finding the Solution

To minimize the cost function, take the derivative with respect to $ \theta $ and set it to zero:

$$
\frac{\partial MSE}{\partial \theta} = -\frac{2}{m} x^T (y - x\theta) = 0
$$

Rearranging gives:

$$
x^T y = x^T x \theta
$$

Finally, solving for $ \theta $ yields the solution:

$$
\theta = (x^T x)^{-1} x^T y
$$

## Implement the Solution

Now that we have proven the solution, implement it in Python.

**Note:**  Make sure to use NumPy for matrix operations.

```python
class LinearRegression:
    def __init__(self) -> None:
        """Initializes the Linear Regression model."""
        self.theta: np.ndarray = None  # Parameters (weights) will be initialized in fit()

    def fit(self, X: np.ndarray, y: np.ndarray) -> None:
        """
        Fits the linear regression model to the training data.

        Args:
            X (np.ndarray): Feature matrix of shape (m, n) where m is the number of samples
                            and n is the number of features.
            y (np.ndarray): Target vector of shape (m, 1) where m is the number of samples.
        """
        ######################  TODO  ########################
        X_bias = np.c_[np.ones((X.shape[0], 1)), X]  
        
        X_transpose = X_bias.T  
        self.theta = np.linalg.inv(X_transpose @ X_bias) @ X_transpose @ y 
        ######################  TODO  ########################

    def predict(self, X: np.ndarray) -> np.ndarray:
        """
        Makes predictions using the learned parameters.

        Args:
            X (np.ndarray): Feature matrix of shape (m, n) where m is the number of samples
                            and n is the number of features.

        Returns:
            np.ndarray: Predicted values of shape (m, 1).
        """
        ######################  TODO  ########################
        X_bias = np.c_[np.ones((X.shape[0], 1)), X] 
        return X_bias @ self.theta
        ######################  TODO  ########################
```

# Minimizing Mean Squared Error (MSE) to Find the Center of a Dataset

In this notebook, we will go through the steps of finding the center of a dataset by minimizing the Mean Squared Error (MSE).

## 1. Formulate the MSE Function

The goal is to find the center \((c_x, c_y)\) that minimizes the MSE between each data point and the center. The MSE function is given by:

$$
MSE(c_x, c_y) = \frac{1}{n} \sum_{i=1}^{n} \left( (x_i - c_x)^2 + (y_i - c_y)^2 \right)
$$

Where:
- \( n \) is the number of data points,
- \( (x_i, y_i) \) are the coordinates of the \(i\)-th data point,
- \( (c_x, c_y) \) is the center we are trying to find.

The objective is to minimize the squared differences between each data point and the center.

## 2. Take the Derivatives of the MSE Function

To find the optimal values of ${c_x}$ and ${c_y}$, we take the partial derivatives of the MSE with respect to ${c_x}$ and ${c_y}$, and set them equal to zero.

### Partial Derivative with Respect to ${c_x}$:

$$
\frac{\partial MSE}{\partial c_x} = \frac{2}{n} \sum_{i=1}^{n} (x_i - c_x)
$$

### Partial Derivative with Respect to ${c_y}$:

$$
\frac{\partial MSE}{\partial c_y} = \frac{2}{n} \sum_{i=1}^{n} (y_i - c_y)
$$

By setting these partial derivatives equal to zero, we get:

$$
\sum_{i=1}^{n} (x_i - c_x) = 0 \quad \text{and} \quad \sum_{i=1}^{n} (y_i - c_y) = 0
$$

### Solving for ${C_x}$ and ${C_y}$:

$$
c_x = \frac{1}{n} \sum_{i=1}^{n} x_i
\quad \text{and} \quad
c_y = \frac{1}{n} \sum_{i=1}^{n} y_i
$$

Thus, the center \((c_x, c_y)\) is simply the **mean** of the \(x\)-coordinates and the **mean** of the \(y\)-coordinates.


```python
######################  TODO  ########################

def find_center(X: np.ndarray, Y : np.ndarray) -> tuple:
    c_x = np.mean(X)
    c_y = np.mean(Y)
    return c_x, c_y

######################  TODO  ########################
```
